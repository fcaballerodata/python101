```python
# OPERACIONES BÁSICAS
ventas_enero = 45000000
ventas_febrero = 52000000

# Suma
total_ventas = ventas_enero + ventas_febrero
print("Total ventas:", total_ventas)  # 97000000

# Resta
diferencia = ventas_febrero - ventas_enero
print("Crecimiento:", diferencia)  # 7000000

# Multiplicación
ventas_trimestre = ventas_enero * 3
print("Proyección trimestre:", ventas_trimestre)

# División
promedio = total_ventas / 2
print("Promedio:", promedio)  # 48500000.0

# División entera (sin decimales)
clientes = 100
dias = 7
clientes_por_dia = clientes // dias
print("Clientes por día:", clientes_por_dia)  # 14

# Módulo (residuo de la división)
residuo = 100 % 7
print("Residuo:", residuo)  # 2

# Potencia
crecimiento = 1.05 ** 12  # 5% mensual durante 12 meses
print("Factor de crecimiento anual:", crecimiento)
```

```python
# Calcular total de una compra (sum())
precios = [10.50, 25.30, 5.99, 12.40, 7.80]
total = sum(precios)

print(f"Total a pagar: ${total}")  # Ese valor tiene muchos decimales
print(f"Total a pagar: ${round(total,2)}")  #Uso de round() para redondear

# Encontrar el Producto Más Caro y Más Barato (max() y min())
precios = [10.50, 25.30, 5.99, 12.40, 7.80]
mas_caro = max(precios)
mas_barato = min(precios)

print(f"Producto más caro: ${round(mas_caro,2)}")
print(f"Producto más barato: ${round(mas_barato,2)}")

# Promedio
promedio =  total / len(precios)
print("El valor promedio es: ", promedio)
```

```python
# sort() modifica la lista original
# sorted() devuelve una nueva lista ordenada sin cambiar la original.
# Usando sorted() (no modifica la lista original)
precios_original = [15, 5, 20, 10, 25]
print(f"Usamos sorted() sobre la lista original: {sorted(precios_original)}")
print(f"Ahora revisemos la lista original a ver si cambió: {precios_original}")
print("La lista original no se modificó")
```

# DÍA 2: Sentencias Condicionales

## Operadores de Comparación

```python
venta_mes_cedri = 410000000
venta_mes_cll80 = 330000000

# == (igual a)
print(venta_mes_cll80 == venta_mes_cedri)
print(venta_mes_cll80 == 330000000)

# != (diferente de)
print(venta_mes_cedri != 150000000)

# > (mayor que)
print(venta_mes_cll80 > venta_mes_cedri)

# < (menor que)
print(venta_mes_cll80 < venta_mes_cedri)

# >= (mayor o igual)
print(venta_mes_cll80 >= venta_mes_cedri)

# <= (mayor que)
print(venta_mes_cll80 <= venta_mes_cedri)

```

## PARTE 2: Estructura IF + ELSE

```python
# Sintaxis básica del IF
ventas = 55000000
meta = 50000000

if ventas >= meta:
    print("✅ ¡Felicitaciones! Cumpliste la meta")
    print(f"Superaste por ${ventas - meta:,}")

# Si la condición es False, no ejecuta nada
```

```python
# IF con ELSE:
ventas = 45000000
meta = 50000000

if ventas >= meta:
    print("✅ Meta cumplida")
else:
    print("❌ No se cumplió la meta")
    faltante = meta - ventas
    print(f"Faltaron ${faltante:,}")
```

## PARTE 4: Operadores Lógicos

```python
# AND - Todas las condiciones deben ser True:

ventas = 55000000
margen = 0.35
nps = 85

mes_excelente = (ventas > 50000000) and (margen > 0.30) and (nps > 80)

if mes_excelente:
    print("🎉 Mes excepcional en todos los indicadores")
else:
    print("📊 Revisar indicadores")
```

```python
# OR - Al menos una condición debe ser True:

ventas = 42000000
nuevos_clientes = 150

requiere_atencion = (ventas < 40000000) or (nuevos_clientes < 100)

if requiere_atencion:
    print("⚠️ Clínica requiere atención")
else:
    print("✅ Operación normal")
```

```python
# NOT - Invierte el resultado:

sistema_actualizado = True

if not sistema_actualizado:
    print("⚠️ Por favor actualiza el sistema")
else:
    print("✅ Sistema OK")
```

## PROYECTO PRÁCTICO Día 2

```python
# Sistema de Evaluación de Performance de Clínica
print("=" * 60)
print("    SISTEMA DE EVALUACIÓN DE PERFORMANCE - CLÍNICA")
print("=" * 60)

# 1. RECOLECTAR DATOS
nombre_clinica = input("\nNombre de la clínica: ")
ventas = float(input("Ventas del mes (COP): "))
costos = float(input("Costos del mes (COP): "))
meta_ventas = float(input("Meta de ventas (COP): "))
clientes = int(input("Número de clientes atendidos: "))
nps = int(input("NPS del mes (0-100): "))

# 2. CALCULAR MÉTRICAS
margen = ((ventas - costos) / ventas) * 100
cumplimiento = (ventas / meta_ventas) * 100
ticket_promedio = ventas / clientes

# 3. CLASIFICACIÓN DE PERFORMANCE
if cumplimiento >= 110 and margen >= 35 and nps >= 80:
    performance = "🌟 EXCELENTE"
    color = "verde"
elif cumplimiento >= 100 and margen >= 30:
    performance = "✅ BUENO"
    color = "azul"
elif cumplimiento >= 90 and margen >= 25:
    performance = "⚠️ REGULAR"
    color = "amarillo"
else:
    performance = "❌ CRÍTICO"
    color = "rojo"

# 3.1 EFICIENCIA OPERATIVA
if margen >= 35 and clientes >= 1000:
    Eficiency = "🌟 EXCELENTE"
    color = "verde"
elif margen >= 30 and clientes >= 800:
    Eficiency = "✅ BUENO"
    color = "azul"
else:
    Eficiency = "⚠️ REGULAR"
    color = "rojo"

# 4. GENERAR ALERTAS
print("\n" + "=" * 60)
print(f"         REPORTE - {nombre_clinica.upper()}")
print("=" * 60)

print(f"\n📊 PERFORMANCE GENERAL: {performance}")
print(f"   Cumplimiento: {cumplimiento:.1f}%")
print(f"   Margen: {margen:.1f}%")
print(f"   NPS: {nps}")
print(f"   Ticket promedio: ${ticket_promedio:,.0f}")
print(f"   Eficiency: {Eficiency}")


print("\n🚨 ALERTAS:")

if   cumplimiento < 90:
     print("   ⚠️ Cumplimiento bajo - Revisar estrategia de ventas")

if   margen < 25:
     print("   ⚠️ Margen bajo - Revisar estructura de costos")

if   nps < 70:
     print("   ⚠️ NPS bajo - Mejorar experiencia del cliente")

if   ticket_promedio < 60000:
     print("   ❌ Ticket bajo - Oportunidad de upselling")
elif ticket_promedio >= 60000 and ticket_promedio <100000:
     print("   ⚠️ Ticket Medio - Oportunidad de crossselling")
elif ticket_promedio >= 100000:
     print("   🌟 Ticket Alto - Documentar buenas prácticas")
else:
     print("    Revisar performance")

# Si no hay alertas
if cumplimiento >= 90 and margen >= 25 and nps >= 70 and ticket_promedio >= 60000:
    print("   ✅ Sin alertas - Operación óptima")

# 5. RECOMENDACIONES
print("\n💡 RECOMENDACIONES:")

if performance == "🌟 EXCELENTE":
    print("   • Documentar mejores prácticas")
    print("   • Compartir estrategias con otras clínicas")
elif performance == "✅ BUENO":
    print("   • Mantener el ritmo actual")
    print("   • Buscar oportunidades de mejora marginal")
elif performance == "⚠️ REGULAR":
    print("   • Enfoque en el indicador más débil")
    print("   • Implementar plan de acción inmediato")
else:  # CRÍTICO
    print("   • Intervención urgente requerida")
    print("   • Reunión con dirección")
    print("   • Plan de recuperación a 30 días")

print("\n" + "=" * 60)
```

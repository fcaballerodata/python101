## Condicionales `if`, `else`, `elif`

Operadores condicionales.png

Los operadores de comparación de Python son palabras clave especiales que se usan para comparar valores y devolver un resultado booleano (`True` o `False`)

```python
print(f"El resultado de 10 > 5 es: {10 > 5}")
print(f"El resultado de 10 < 5 es: {10 < 5}")
print(f"El resultado de 10 >= 5 es: {10 >= 5}")
print(f"El resultado de 10 <= 5 es: {10 <= 5}")
print(f"El resultado de 10 == 5 es: {10 == 5}")
print(f"El resultado de 10 != 5 es: {10 != 5}")

print(f"El resultado de 7 == 7.0 es: {7 == 7.0}")
print(f"El resultado de 'casa' != 'perro' es: {'casa' != 'perro'}")
print(f'El resultado de "manzana" == "pera" es: {"manzana" == "pera"}')
print(f'El resultado de "manzana" == "manzana" es: {"manzana" == "manzana"}')
print(f'El resultado de "beta" > "alfa" es: {"beta" > "alfa"}')
print(f'El resultado de "beta" < "alfa" es: {"beta" < "alfa"}')

print(f"El resultado de [1, 2, 3] == [1, 2, 3] es: {[1, 2, 3] == [1, 2, 3]}")
print(f"El resultado de [1, 2, 3] == (1, 2, 3) es: {[1, 2, 3] == (1, 2, 3)}")

print(f"El resultado de True == 1 es: {True == 1}")
print(f"El resultado de False == 0 es: {False == 0}")
print(f"El resultado de False == '' es: {False == ''}")
```

```python
edad = 78

if edad >= 19: # if + condición + :
    print("Eres mayor de edad.") #Atención a la identación

else: # No lleva condición porque es la "alternativa" al if
    print("Eres menor de edad.")
```

## El Bucle While
* Se utiliza cuando quieres repetir un ciclo mientras se cumpla una condición.

```python
# Contar del 1 al 5

contador = 1

while contador <= 5:
    print("Contador:", contador)
    contador += 1  # aumenta en 1 para que el ciclo termine

print("Fin del ciclo")



```

## IF, ELSE + Operadores + metodos

📌 Ejemplo: Validar si un código de usuario cumple con estas reglas:

* Debe contener al menos 1 caracter en mayuscula
* Debe tener al menos un número o caracter especial
* Debe tener al menos 5 caracteres

```python
codigo_usuario="User_23"

print(codigo_usuario.isalpha()) #No acepta caracteres especiales ni numeros
print(codigo_usuario.islower()) #Valida que todos las letras estén en minusculas, ignora otros caracteres y numeros
print(codigo_usuario.isdigit()) #No acepta caracteres especiales ni letras
```

```python
password = "Segura"

if not password.islower() and not password.isalpha() and len(password) > 6:
    print("Contraseña válida")
else:
    print("La contraseña debe tener al menos un caracter en mayuscula, un numero o caracter especial y tener más de 6 caracteres.")
```

### CONDICIONALES ANIDADOS ¿Como ser mas específicos en el error de la contraseña?

Se utilizan para validar condiciones secuenciales, dependientes... es decir, si una no se cumple, no se validan las otras.

```python
password = "Segura_123"

if not password.islower():
    print("La contraseña SI tiene un caracter en mayuscula.")

    if not password.isalpha():
        print("La contraseña SI tiene un caracter especial o número.")

        if len(password)>6:
            print("La contraseña SI tiene más de 6 caracteres.")
            print("Contraseña válida")
        else:
            print("La contraseña NO tiene más de 6 caracteres.")
    else:
        print("La contraseña NO tiene un caracter especial o número.")
else:
    print("La contraseña NO tiene un caracter en mayuscula.")


```

```python
passwords = ["Segura_123", "micedula@#", "mayito123", "YoSoyAndres_9858","MiNo1"]

for password in passwords:
  print(f"\nValidando contraseña: {password}   ---------* ")
  if not password.islower():
      print("La contraseña SI tiene un caracter en mayuscula.")

      if not password.isalpha():
          print("La contraseña SI tiene un caracter especial o número.")

          if len(password)>6:
              print("La contraseña  SI tiene más de 6 caracteres.")
              print(f"Contraseña {password} es válida")
          else:
              print(f"La contraseña {password} NO tiene más de 6 caracteres.")
      else:
          print(f"La contraseña {password} NO tiene un caracter especial o número.")
  else:
      print(f"La contraseña  {password} NO tiene un caracter en mayuscula.")
```

## CONDICIONALES ELIF:

*   Se usan cuando hay opciones excluyentes (solo una se cumple).
*   Si una es True, las demás se ignoran.


```python
numero = [5,-8,4,0, 5.5,0, -15,-4,0] # Prueba con diferentes valores

for number in numero:
  if number > 0:
      print("El número es positivo.")
  elif number == 0:
      print("El número es cero.")
  else:
      print("El número es negativo.")
```

```python
rol = [["Jorge" ,"admin"],["Laura", "editor"], ["Camilo","usuario"], ["Ana", "visitante"]] # Prueba con diferentes valores

for user in rol:
  if user[1] == "admin":
      print(f"{user[0]} tienes acceso total concedido.")
  elif user[1] == "editor":
      print(f"{user[0]} tienes acceso parcial: puedes editar contenido.")
  elif user[1] == "usuario":
      print(f"{user[0]} tienes acceso limitado: solo puedes ver contenido.")
  else:
      print(f"Acceso denegado.")
```

```python
# Lista de estudiantes: [Nombre, Carrera, Beca asignada (Sí/No)]
estudiantes = [
    ["Ana Pérez", "Ingeniería", "No"],
    ["Carlos Gómez", "Medicina", "No"],
    ["Luis Martínez", "Derecho", "No"],
    ["María López", "Ingeniería", "No"],
    ["Jorge Ramírez", "Medicina", "No"],
    ["Lucía Torres", "Arquitectura", "No"],
    ["Pedro Sánchez", "Ingeniería", "No"]
]

# Definimos el presupuesto y la cantidad de becas disponibles
presupuesto = 10000  # Dinero total disponible


indice = 0

# Mientras haya becas y dinero, seguimos asignando
while  presupuesto > 0 and indice < len(estudiantes):
    estudiante = estudiantes[indice]
    carrera = estudiante[1]

    # Asignamos el costo de beca manualmente
    if carrera == "Ingeniería":
        costo = 2500
    elif carrera == "Medicina":
        costo = 5000
    elif carrera == "Derecho":
        costo = 2000
    elif carrera == "Arquitectura":
        costo = 2200
    else:
        costo = 0  # Si la carrera no está en la lista, no se asigna beca

    # Si hay suficiente presupuesto, se asigna la beca
    if presupuesto >= costo and costo > 0:
        estudiante[2] = "Sí"
        presupuesto -= costo
        print(f"✅ Beca asignada a {estudiante[0]} en {carrera}. Presupuesto restante: ${presupuesto}")

    indice += 1  # Pasamos al siguiente estudiante

# Mostramos los resultados
print("Lista de estudiantes con becas asignadas:")
for estudiante in estudiantes:
    print(estudiante)

print("Presupuesto restante:", presupuesto)
```
